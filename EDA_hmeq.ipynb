****3.4 Distribution****

3.4.1. Continuous

```python
num_columns = ['LOAN', 'MORTDUE', 'VALUE', 'YOJ', 'CLAGE', 'DEBTINC']
num_rows = 3
num_cols = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 15))

axes = axes.flatten()

for idx, column in enumerate(num_columns):
    ax = axes[idx]

    ax.hist(data_train[column].dropna(), bins=20, range=(0, data_train[column].max()), edgecolor='black')

    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')
    ax.set_title(f'Histogram of {column}')

    ax.grid(False)

plt.tight_layout()
plt.show()
```

3.4.2. Discrete

```python
import matplotlib.pyplot as plt

discrete_variables = ['DEROG', 'DELINQ', 'NINQ', 'CLNO']

num_rows = 2
num_cols = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 10))

axes = axes.flatten()

for i, var in enumerate(discrete_variables):
    ax = axes[i]
    data_train[var].value_counts().sort_index().plot(kind='bar', color='blue', ax=ax)
    ax.set_xlabel(f'{var} Values')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Bar Chart of {var} Values')
    ax.grid(False)

if len(discrete_variables) < num_rows * num_cols:
    for j in range(len(discrete_variables), num_rows * num_cols):
        axes[j].axis('off')

plt.tight_layout()
plt.show()
```

****3.5 Data Trend****

```python
# Function to add count labels
def add_count_labels(ax):
    for p in ax.patches:
        count = int(p.get_height())
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(count, (x, y), ha='center', va='bottom')

# Create subplots for bar plots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
fig.suptitle("Relationships between 'BAD' and Different Features", fontsize=16)

# Bar plot for 'REASON'
ax1 = sns.countplot(x='REASON', hue='BAD', data=data_train, ax=axes[0])
axes[0].set_title("Relationship between 'BAD' and 'REASON'")
axes[0].set_xlabel('Reason')
axes[0].set_ylabel('Count')
axes[0].legend(title='BAD')
add_count_labels(ax1)

# Bar plot for 'JOB'
ax2 = sns.countplot(x='JOB', hue='BAD', data=data_train, ax=axes[1])
axes[1].set_title("Relationship between 'BAD' and 'JOB'")
axes[1].set_xlabel('Job Category')
axes[1].set_ylabel('Count')
axes[1].legend(title='BAD')
add_count_labels(ax2)

# Adjust layout
fig.tight_layout()
plt.show()
```

3.6.2 Skewness

```python
from scipy.stats import skew

columns = ["MORTDUE", "VALUE", "YOJ", "CLAGE", "LOAN"]

skewed_feats = data_train[columns].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness.head(15)
```

```python
from scipy.stats import norm, skew, probplot
data_train = data_train.dropna()  # Loại bỏ các dòng có giá trị NaN


for col in columns:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Histogram
    sns.histplot(data_train[col], kde_kws={'fit': norm}, ax=axes[0])
    (mu, sigma) = norm.fit(data_train[col])
    axes[0].legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)], loc='best')
    axes[0].set_ylabel('Frequency')
    axes[0].set_title(f'{col} Distribution')

    # QQ-plot
    probplot(data_train[col], plot=axes[1])
    axes[1].set_title(f'QQ Plot for {col}')

    plt.tight_layout()
    plt.show()
```
