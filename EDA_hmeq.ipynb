***2.Target value***

```python
data['BAD'].value_counts()
```

```python
from matplotlib import pyplot as plt

data['BAD'].value_counts().plot(kind='bar', color='blue')
plt.xlabel('BAD Values')
plt.ylabel('Frequency')
plt.title('Bar Chart of BAD Values')
plt.show()
```
