**1. Generate CCCD and Name**

```python
import pandas as pd
import random

file_path = 'D:\\WorkSpace_Thinh1\\FSS\\hmeq_with_scores.csv'
data = pd.read_csv(file_path)

def generate_cccd():
    return ''.join([str(random.randint(0, 9)) for _ in range(12)])

last_names = ['Nguyễn', 'Trần', 'Lê', 'Phạm', 'Hoàng', 'Huỳnh', 'Phan', 'Vũ', 'Đặng', 'Bùi']
middle_names = ['Văn', 'Thị', 'Hồng', 'Quốc', 'Thùy', 'Thu', 'Minh', 'Đức', 'Xuân', 'Hải']
first_names = ['Anh', 'Bình', 'Cường', 'Dũng', 'Hà', 'Hùng', 'Lan', 'Linh', 'Mai', 'Phương', 'Quyên', 'Sơn', 'Thảo', 'Tú', 'Vân']

def generate_name():
    return f"{random.choice(last_names)} {random.choice(middle_names)} {random.choice(first_names)}"


data['CCCD'] = [generate_cccd() for _ in range(len(data))]
data['Tên'] = [generate_name() for _ in range(len(data))]




```

**2. Generate HDV (Mã hợp đồng vay)**

```python
import pandas as pd
import random
import string

file_path = 'D:\\WorkSpace_Thinh1\\FSS\\hmeq_with_scores.csv'
data = pd.read_csv(file_path)

def generate_unique_contract_id(existing_ids):
    while True:
        contract_id = ''.join(random.choices(string.digits, k=6))
        if contract_id not in existing_ids:
            existing_ids.add(contract_id)
            return contract_id

existing_ids = set()

data['HDV'] = [generate_unique_contract_id(existing_ids) for _ in range(len(data))]


```
